```python
from matplotlib import pyplot as plt 
import numpy as np
import pandas as pd 
import plotly.express as px
import datetime
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
df = pd.read_csv( '/content/final_df.csv')

# df = pd.read_csv( '/content/vms_eureka_trades_all_day.csv') 


stock = pd.read_csv('/content/bank_nifty_stocks.csv')
bank_nifty = pd.read_csv('/content/bank_nifty_index_data_eod.csv')
```

```python
# print(df.columns)

# print(stock.columns)
# # print(stock.head())
```

```python

bank_nifty_list = ['HDFCBANK' , 'ICICIBANK' , 'KOTAKBANK', 'SBIN' , 'AXISBANK', 'INDUSINDBK', 'BANKBARODA' ,'IDFCFIRSTB','FEDERALBNK' ,'PNB'] 


stock['total_delivery_turnover'] =  0 

stock['total_turnover'] =  0 

stock['gapups'] = 0 
stock['gadowns'] = 0 


for i in bank_nifty_list:
    print(i)

    
    ### turnover ###
    stock[i + '_delivery_turnover' ] =  stock[ i + '_deliverable_volume'] *  stock[ i + '_VWAP']

    stock['total_delivery_turnover'] =  stock['total_delivery_turnover'] + stock[i + '_delivery_turnover' ]

    stock['total_turnover'] =  stock['total_turnover'] + stock[i + '_turnover' ]

    #### gap up gap down #### 

    stock[i+ '_gapup%'] = ( stock[ i + '_open'] - stock[ i + '_prev_close']) /  stock[ i + '_prev_close']  *100 



 


    



stock['total_delivery_percentage'] =   stock['total_delivery_turnover'] / stock['total_turnover']
    


for i in bank_nifty_list:
    print(i)
    stock[i+'_open_above_high'] = 0


    for j in range( 1 , len(stock) ): 


      if stock[i+ '_open'].iloc[j] > stock[i + '_high'].iloc[j -1 ]  or stock[i+ '_open'].iloc[j] < stock[i + '_low'].iloc[j -1 ] :

        stock[i+'_open_above_high'].iloc[j] = 1 
      
    


```

```python
# df1
df.rename(columns={ 'Entry Date' : 'Date' } , inplace = True) 
df['Date']= pd.to_datetime(df['Date'])
df['Weekday'] = df['Date'].dt.day_name()
df.set_index( 'Date',inplace = True ) 

# df2
bank_nifty['Date']= pd.to_datetime(bank_nifty['Date'])
bank_nifty.set_index( 'Date',inplace = True )

bank_nifty['open_above_high'] = 0
bank_nifty['prev_day_high-low'] = 0 
bank_nifty['prev_day_close-open'] = 0 
bank_nifty['prev_day_volume'] = 0  

bank_nifty['prev_day_close-open_by_volume'] = 0 
bank_nifty['prev_day_turnover']  = 0 
 


for j in range( 1 , len(bank_nifty) ): 

      bank_nifty['prev_day_close-open'].iloc[j]  = (bank_nifty['Close'].iloc[j -1 ] - bank_nifty['Open'].iloc[j -1 ]) / bank_nifty['Close'].iloc[j -1 ]
      bank_nifty['prev_day_high-low'].iloc[j]  = (bank_nifty['High'].iloc[j -1 ] - bank_nifty['Low'].iloc[j -1 ]) / bank_nifty['Low'].iloc[j -1 ]
      
      bank_nifty['prev_day_volume'].iloc[j]  = (bank_nifty['Volume'].iloc[j -1 ])   
      
      bank_nifty['prev_day_turnover'].iloc[j]  = (bank_nifty['Turnover'].iloc[j -1 ])   
 

      bank_nifty['prev_day_close-open_by_volume'].iloc[j]  = (bank_nifty['prev_day_close-open'].iloc[j -1 ] / bank_nifty['Volume'].iloc[j -1 ])


      if bank_nifty['Open'].iloc[j] > bank_nifty['High'].iloc[j -1 ] or True  :

        bank_nifty['open_above_high'].iloc[j] =    (bank_nifty['Open'].iloc[j]  - bank_nifty['Close'].iloc[j -1 ]) /  bank_nifty['Close'].iloc[j -1 ]

      



new_df= pd.concat([df, bank_nifty], axis=1 , join = 'inner' )




print(new_df.head())



```
